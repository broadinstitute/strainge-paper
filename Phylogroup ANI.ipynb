# Phylogroup ANI

```python
from IPython.display import display
import pandas

ani = pandas.read_csv('cobra2/ANI.tsv', sep='\t')
ani['genome1'] = ani['#genome1:genome2'].str.split(':', expand=True)[0]
ani['genome2'] = ani['#genome1:genome2'].str.split(':', expand=True)[1]
ani = ani.set_index(['genome1', 'genome2'])
```

```python
import json

with open("clades.json") as f:
    clades = json.load(f)
    
ani['clade1'] = ani['genome1'].map(clades)
ani['clade2'] = ani['genome2'].map(clades)

ani_same_clade = ani[ani['clade1'] == ani['clade2']]

ani_same_clade.groupby(['clade1']).mean()
```
